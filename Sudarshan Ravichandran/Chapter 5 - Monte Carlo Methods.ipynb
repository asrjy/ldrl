## Prediction and Control Tasks

in rl, there are two importatnt tasks. 

- prediction task
- control task

in prediction task, we are are given a policy. we tell the expected return we get from using that policy. we do not make any changes to the policy given.

in control task, we start off with a random policy and we iteratively find out the optimal policy.
